```python
df_country=pd.read_excel("Country-Code.xlsx")
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_name=final_df.Country.value_counts().index
country_value=final_df.Country.value_counts().values
```

```python
plt.pie(country_value[:3],labels=country_name[:3],autopct='%1.2f%%')
```

**Obersvation:Zomato maximum records or transaction are from India ,USA,UK and followed by other countries respectively**

```python
ratings=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"rating count"})
```

```python
ratings
```

**Observation : ...................
1: when rating is between 4.5 to 4.9 ---> Excellent
2: when rating is between 4.0 to 4.4 ---> Very good
3: when rating is between 3.5 to 3.9 ---> Good
4: when rating is between 2.5 to 3.4 ---> Average
5: when rating is between 1.8 to 2.4 ---> Poor
6: when rating is between 0 ---> Not Given**

```python
ratings.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating", y="rating count",data=ratings)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating", y="rating count",data=ratings,hue="Rating color",palette=["white","red","orange","yellow","green","green"])
```

```python
#**Observation:..........
#1: not rated count is very high
#2: maximun number of ratings are between 2.5 and 3.4**
```

```python
sns.countplot(x="Rating color",data = ratings ,palette= ["blue","red","orange","yellow","green","green"])
```
